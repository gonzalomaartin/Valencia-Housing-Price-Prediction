```python
import numpy as np 
import pandas as pd, ast
import re 
```

```python
df = pd.read_csv("../working_data/properties_vlc.csv")
```

```python
df.head()
```

```python
df["location"] = df["location"].apply(ast.literal_eval)
```

```python
df[["street", "area", "city"]] = pd.DataFrame(df["location"].apply(lambda x: x[-3:] if len(x) >= 3 else [np.nan]*(3-len(x)) + x).tolist(), index=df.index)
df[["street", "area", "city"]] = df[["street", "area", "city"]].replace("", np.nan)
```

```python
df.loc[df["city"].isna(), ["street", "area", "city"]].head()
```

```python
street_mean = df[df["property_age"] == 2025].groupby("street")["property_age"].mean()
area_mean = df[df["property_age"] == 2025].groupby("area")["property_age"].mean()
city_mean = df[df["property_age"] == 2025].groupby("city")["property_age"].mean()
```
